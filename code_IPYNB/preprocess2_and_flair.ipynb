```
# download flair library #
import torch
import flair
```

```
import pandas as pd
df=pd.read_csv("train.csv")
```

```
from flair.embeddings import WordEmbeddings
from flair.embeddings import CharacterEmbeddings
from flair.embeddings import StackedEmbeddings
from flair.embeddings import FlairEmbeddings
from flair.embeddings import BertEmbeddings
from flair.embeddings import ELMoEmbeddings
from flair.embeddings import FlairEmbeddings

### Initialising embeddings (un-comment to use others) ###
#glove_embedding = WordEmbeddings('glove')
#character_embeddings = CharacterEmbeddings()
flair_forward  = FlairEmbeddings('news-forward-fast')
flair_backward = FlairEmbeddings('news-backward-fast')
#bert_embedding = BertEmbedding()
#elmo_embedding = ElmoEmbedding()
stacked_embeddings = StackedEmbeddings( embeddings = [ 
                                                       flair_forward, 
                                                       flair_backward
                                                      ])
```

```
df['para']='nan'
```

```
for i in range(len(df['statement'])):
  df['para'][i]=str(df['statement'][i])+str(df['justification'])
```

```
df['para'][0]
```

```
#Data Cleaning for Train, Val, Test
text = df['para'] 

txt = text.tolist()
for i in range(0,len(txt)):
    txt[i] = str(txt[i]).replace('.*([0-9]\.[0-9]+)',"")
    
for i in range(0,len(txt)):
    txt[i] = txt[i].lower()
    
import re
for i in range(0,len(txt)):
    txt[i] = txt[i].replace('\d+', '')
    
import string
for i in range(0,len(txt)):
    txt[i] = re.sub(r'[^\w\s]','',txt[i])
for i in range(0,len(txt)):
    txt[i] = re.sub(r'[^\w\s]','',txt[i])
for i in range(0,len(txt)):
    txt[i] = txt[i].strip()
    
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('english')

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
for i in range(0, len(txt)):
    tx = word_tokenize(txt[i])
    new_txt = []
    for word in tx:
        new_txt.append(lemmatizer.lemmatize(word))
    txt[i] = ' '.join(word for word in new_txt)
    
from string import digits 
def remove(liste):
    remove_digits = str.maketrans('', '', digits) 
    liste = [i.translate(remove_digits) for i in liste] 
    return liste
txt = remove(txt)

from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
nltk.download('stopwords')
for i in range(0, len(txt)):
    stop_words = set(stopwords.words('english')) 
    word_tokens = word_tokenize(txt[i]) 
    filtered_sentence = [] 
    for w in word_tokens: 
        if w not in stop_words: 
            filtered_sentence.append(w) 
    txt[i] = txt[i] = ' '.join(word for word in filtered_sentence)
df['new_txt'] = txt
```

```
df['new_txt']
```

```
from flair.data import Sentence
# create a sentence #
sentence = Sentence(df['new_txt'][0])
# embed words in sentence #
stacked_embeddings.embed(sentence)
for token in sentence:
    print(token.embedding)
# data type and size of embedding #
print(type(token.embedding))
# storing size (length) #
z = token.embedding.size()[0]
```

```
from tqdm import tqdm ## tracks progress of loop ##

# creating a tensor for storing sentence embeddings #
s = torch.zeros(0,z)

# iterating Sentence (tqdm tracks progress) #

for tweet in tqdm(df['new_txt']):
    # empty tensor for words #
    w = torch.zeros(0,z)   
    sentence = Sentence(tweet)
    stacked_embeddings.embed(sentence)
  # for every word #
    for token in sentence:
        # storing word Embeddings of each word in a sentence #
        w = torch.cat((w,token.embedding.view(-1,z)),0)
  # storing sentence Embeddings (mean of embeddings of all words)   #
    s = torch.cat((s, w.mean(dim = 0).view(-1, z)),0)
```

```
X = s.numpy()
```

```
from pandas import DataFrame
kane = DataFrame.from_records(X)
kane
```
