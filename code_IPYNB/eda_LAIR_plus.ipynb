```python
labels = 'half-true', 'false', 'mostly-true', 'barely-true', 'true', 'pants-fire'
sizes = [265, 249, 241, 212, 208, 92]
colors = ['B', 'silver', 'G', 'yellow', '#A00994', '#E40E00']
explode = (0.1, 0.0, 0.1, 0, 0.1, 0.0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title("Percentage of Different Types of Labels")
plt.plot()
fig=plt.gcf()
fig.set_size_inches(5,5)
plt.show()
```

```python
descending_order = df["subject"].value_counts()[:20].sort_values(ascending=False).index
fig,ax=plt.subplots()
fig.set_size_inches(20,10)
ax = sn.countplot(x="subject", data=df,order=descending_order)
xticks(rotation=90)
```
